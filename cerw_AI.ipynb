```python
author_agent = Agent(
    role="Author",
    goal= "generate stories given some details about it",
    backstory= "german author with 5 years experince who loved and inspired by ahmed khalid tawfig ",
    llm=llm,
    verbose=True


)

genertae_task= Task(
    description = "Generate Story about {person_name} and about {subject} for non native english speakers",
    expected_output= "maximum ten lines of story",
    agent=author_agent



)
```

```python
summrizer_agent = Agent(
    role="summerizer Author",
    goal= "summrize stories ",
    backstory= "american author with 5 years experince  ",
    llm=llm,
    verbose=True


)

summarize_task= Task(
    description = "summrize Story ",
    expected_output= "summrized story with maximum 100 words",
    agent=summrizer_agent



)
```

```python
crew = Crew(
    agent = [
        author_agent,
        summrizer_agent
    ],
    tasks = [
        genertae_task,
        summarize_task
    ],
    process = Process.sequential







)
```

```python
result = crew.kickoff(
    inputs = {
        "person_name" : "dr ahmed",
        "subject":"his career"

    }




)
```
